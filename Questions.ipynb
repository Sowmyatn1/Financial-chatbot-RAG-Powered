```python
def simple_chatbot(user_query):
   if user_query == "What is the total revenue of Apple?":
       return "The total revenue is 1168.6B."
   elif user_query == "What is the total revenue of Microsoft?":
       return "The total revenue is 655.3B ."
   if user_query == "What is the total revenue of Tesla?":
       return "The total revenue is 232.1B ."
   elif user_query == "How has net income changed over the last year for Apple?":
       return "The net income has increased by 2% over the last year."
   elif user_query == "How has net income changed over the last year for Microsoft?":
       return "The net income has increased by 15.6% over the last year."
   elif user_query == "How has net income changed over the last year for Tesla?":
       return "The net income has decreased by 18.7 over the last year."
   # Add more conditions for other predefined queries
   else:
       return "Sorry, I can only provide information on predefined queries."
```
