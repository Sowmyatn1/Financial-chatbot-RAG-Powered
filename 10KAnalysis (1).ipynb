```python
total_revenue_by_year = df.groupby('Year')['Total Revenue'].sum().reset_index()

# Merge to calculate percentage contribution of each company
df_with_total_revenue = pd.merge(df, total_revenue_by_year, on='Year', suffixes=('', '_Total'))
df_with_total_revenue['Revenue Percentage'] = (df_with_total_revenue['Total Revenue'] / df_with_total_revenue['Total Revenue_Total']) * 100

# Display the revenue percentage contribution for each company by year
revenue_percentage_by_company_year = df_with_total_revenue[['Company', 'Year', 'Revenue Percentage']]
print(revenue_percentage_by_company_year)
```


#Revenue Contribution by Year (As Percentage of Total Revenue)
Apple contributes the largest percentage of revenue each year, maintaining a steady contribution around 56%.
Microsoft's contribution grows slightly, especially in 2024, and Tesla's contribution is relatively small compared to Apple and Microsoft.

```python
# Calculate Net Income Margin for each company and year
df['Net Income Margin (%)'] = (df['Net Income'] / df['Total Revenue']) * 100

# Display the results
net_income_margin = df[['Company', 'Year', 'Net Income Margin (%)']]
print(net_income_margin)
```
